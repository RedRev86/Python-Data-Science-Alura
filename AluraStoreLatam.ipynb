# 2. Ventas por categoría

```python
# Contar la cantidad de productos vendidos por categoría en la tienda. Por defecto la tienda 1.
ventas_por_categoria = tienda1['Categoría del Producto'].value_counts().reset_index()
ventas_por_categoria.columns = ['Categoría', 'Cantidad de Ventas']
ventas_por_categoria
```

```python
# Lista con los DataFrames y nombres de tiendas
tiendas = [
          (tienda1, "Tienda 1"),
          (tienda2, "Tienda 2"),
          (tienda3, "Tienda 3"),
          (tienda4, "Tienda 4")
          ]

# Calcular ventas por categoría para cada tienda
ventas_por_tienda = []
for df, nombre in tiendas:
    ventas = df['Categoría del Producto'].value_counts().reset_index()
    ventas.columns = ['Categoría', 'Cantidad de Ventas']
    ventas['Tienda'] = nombre
    ventas_por_tienda.append(ventas)

# Mostrar resultados individuales
for ventas in ventas_por_tienda:
    print(f"\n▶ Ventas por categoría - {ventas['Tienda'].iloc[0]}")
    print(ventas)
```

```python
# Total por categoría
# Unir los DataFrames de las 4 tiendas para obtener el valor total
df_total = pd.concat([tienda1, tienda2, tienda3, tienda4], ignore_index=True)

# Agrupar por categoría y contar la cantidad de ventas
ventas_totales = df_total['Categoría del Producto'].value_counts().reset_index()
ventas_totales.columns = ['Categoría', 'Cantidad de Ventas']

# Mostrar resultado
print("▶ Ventas totales por categoría en todas las tiendas")
print(ventas_totales)

```

# 3. Calificación promedio de la tienda


```python
# Calificación promedio por tienda

calificacion_tienda1 = tienda1['Calificación'].mean()
calificacion_tienda2 = tienda2['Calificación'].mean()
calificacion_tienda3 = tienda3['Calificación'].mean()
calificacion_tienda4 = tienda4['Calificación'].mean()

# Mostrar resultados
print(f"▶ Calificación promedio Tienda 1: {calificacion_tienda1:.2f}")
print(f"▶ Calificación promedio Tienda 2: {calificacion_tienda2:.2f}")
print(f"▶ Calificación promedio Tienda 3: {calificacion_tienda3:.2f}")
print(f"▶ Calificación promedio Tienda 4: {calificacion_tienda4:.2f}")
```

Otra forma de obtener el resultado:

```python
# Lista de tiendas con nombre y DataFrame
tiendas = [
    ("Tienda 1", tienda1),
    ("Tienda 2", tienda2),
    ("Tienda 3", tienda3),
    ("Tienda 4", tienda4),
          ]

# Lista para guardar resultados
resultados = []

for nombre, df in tiendas:
    promedio = df['Calificación'].mean()
    resultados.append({"Tienda": nombre, "Calificación Promedio": promedio})

# Crear DataFrame
df_promedios = pd.DataFrame(resultados)

# Gráfica de barras
plt.figure(figsize=(8, 5))
sns.pointplot(data=df_promedios, x="Tienda", y="Calificación Promedio", palette="viridis")

plt.title("Calificación promedio por tienda")
plt.ylabel("Calificación Promedio")
plt.ylim(0, 5)  # Ajusta si tu escala de calificación es del 1 al 5
plt.grid(axis='y', linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()
```

```python
# Para mejorar la visualización de la gráfica
plt.figure(figsize=(8, 5))
sns.lineplot(data=df_promedios, x="Tienda", y="Calificación Promedio", marker="o", color="teal")

# Zoom en Y
plt.ylim(df_promedios["Calificación Promedio"].min() - 0.1,
         df_promedios["Calificación Promedio"].max() + 0.1)

# Etiquetas en cada punto
for i, row in df_promedios.iterrows():
    plt.text(row["Tienda"], row["Calificación Promedio"] + 0.02,
             f'{row["Calificación Promedio"]:.2f}', ha='center', fontsize=10)

# Línea de referencia (promedio)
media_global = df_promedios["Calificación Promedio"].mean()
plt.axhline(media_global, color='gray', linestyle='--', alpha=0.5, label='Media Global')
plt.legend()

plt.title("Calificación promedio por tienda")
plt.ylabel("Calificación Promedio")
plt.grid(True, linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()
```

# 4. Productos más y menos vendidos

```python
# Función para mostrar productos más y menos vendidos de una tienda
def productos_mas_menos_vendidos(df, nombre_tienda):
    conteo = df['Producto'].value_counts()

    producto_mas = conteo.idxmax()
    ventas_mas = conteo.max()

    producto_menos = conteo.idxmin()
    ventas_menos = conteo.min()

    print(f"📦 {nombre_tienda}")
    print(f"Más vendido: {producto_mas} ({ventas_mas} ventas)")
    print(f"Menos vendido: {producto_menos} ({ventas_menos} venta{'s' if ventas_menos > 1 else ''})\n")

    # Visualización
    top10 = conteo.head(10)
    plt.figure(figsize=(10,4))
    #sns.barplot(x=top10.values, y=top10.index, palette="viridis")
    sns.barplot(x=top10.values, y=top10.index, hue=top10.index, palette="viridis", legend=False)
    plt.title(f"Top 10 productos más vendidos - {nombre_tienda}")
    plt.xlabel("Número de Ventas")
    plt.ylabel("Producto")
    plt.tight_layout()
    plt.show()

# Aplicar para cada tienda
productos_mas_menos_vendidos(tienda1, "Tienda 1")
productos_mas_menos_vendidos(tienda2, "Tienda 2")
productos_mas_menos_vendidos(tienda3, "Tienda 3")
productos_mas_menos_vendidos(tienda4, "Tienda 4")
```

```python
# Crear función para graficar las gráficas de cada una de las tiendas
def graficar_envio_por_producto(df_tienda, nombre_tienda):
    # Ordenar por costo de envío para una línea más clara
    df_ordenado = df_tienda.sort_values('Costo de envío', ascending=True).reset_index(drop=True)

    plt.figure(figsize=(12, 5))
    sns.lineplot(data=df_ordenado, x='Producto', y='Costo de envío', marker='o', color='teal')

    # Marcar el producto con mayor costo
    max_idx = df_ordenado['Costo de envío'].idxmax()
    max_producto = df_ordenado.loc[max_idx, 'Producto']
    max_valor = df_ordenado.loc[max_idx, 'Costo de envío']

    plt.scatter(max_producto, max_valor, color='red', s=100, zorder=5)
    plt.text(max_producto, max_valor + max_valor * 0.03, f'Máximo: {max_valor:,.0f}',
             ha='center', va='bottom', color='red', fontweight='bold')

    plt.title(f'Costo de envío por producto - {nombre_tienda}')
    plt.xlabel('Producto')
    plt.ylabel('Costo de envío')
    plt.xticks(rotation=45, ha='right')
    plt.grid(True, linestyle='--', alpha=0.5)
    plt.tight_layout()
    plt.show()

# Llamadas para cada tienda
graficar_envio_por_producto(tienda1, "Tienda 1")
graficar_envio_por_producto(tienda2, "Tienda 2")
graficar_envio_por_producto(tienda3, "Tienda 3")
graficar_envio_por_producto(tienda4, "Tienda 4")
```

##**Informe Final de Rendimiento de Tiendas**

Este informe presenta un análisis de desempeño de las tiendas con base en ventas, opiniones de clientes, productos y costos de envío.

```python
# Código base de comparación entre tiendas
# Lista con nombre y DataFrame de cada tienda
tiendas = [
          ("Tienda 1", tienda),
          ("Tienda 2", tienda2),
          ("Tienda 3", tienda3),
          ("Tienda 4", tienda4)
          ]

# Lista para el resumen
resumen = []

for nombre, df in tiendas:
    resumen.append({
        "Tienda": nombre,
        "Ingresos": df["Precio"].sum(),
        "Promedio Calificación": df["Calificación"].mean(),
        "Costo Promedio Envío": df["Costo de envío"].mean(),
        "Total Productos Vendidos": df["Producto"].count()
    })

df_resumen = pd.DataFrame(resumen)
df_resumen_sorted = df_resumen.sort_values(by="Ingresos", ascending=False)
display(df_resumen_sorted)
```

```python
# Tienda con menores ingresos
tienda_menor = df_resumen_sorted.iloc[-1]

print("INFORME FINAL\n")
print("Resumen de rendimiento por tienda:")
print(df_resumen_sorted.to_string(index=False))

print("\n Observaciones clave:")
print(f"- La tienda con menores ingresos es {tienda_menor['Tienda']} con ${tienda_menor['Ingresos']:,.0f} en ventas.")
print(f"- Su calificación promedio es {tienda_menor['Promedio Calificación']:.2f} y el costo promedio de envío es de ${tienda_menor['Costo Promedio Envío']:,.0f}.")

if tienda_menor["Ingresos"] < df_resumen_sorted["Ingresos"].mean() * 0.75:
    print(f"\n❗ Recomendación: Se sugiere analizar la viabilidad de mantener {tienda_menor['Tienda']}, ya que sus ingresos están considerablemente por debajo del promedio.")
else:
    print(f"\n✅ {tienda_menor['Tienda']} tiene ingresos cercanos al promedio, se recomienda evaluar su cierre .")
```
